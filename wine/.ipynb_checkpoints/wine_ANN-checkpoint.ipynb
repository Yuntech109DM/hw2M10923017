```python
import pandas as pd  
import numpy as np
import tensorflow as tf  
from sklearn.metrics import explained_variance_score,mean_absolute_error,median_absolute_error
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('WineQuality.data',header=None,encoding='utf-8')
```

```python
df.columns=['fixed', 'volatile', 'citric', 'sugar', 'chlorides','dioxide','td','density','pH','sulphates','alcohol','quality']
```

```python
df.describe().T
```

```python
df.info()
```

```python
X = df[['citric', 'sugar', 'chlorides','density','pH','sulphates','alcohol']]
```

```python
y = df['quality']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=23)
```

```python
X_test, X_val, y_test, y_val = train_test_split(X_test, y_test, test_size=0.5, random_state=23)
```

```python
X_train.shape, X_test.shape, X_val.shape 
```

```python
feature_cols = [tf.feature_column.numeric_column(col) for col in X.columns]
```

```python
regressor = tf.estimator.DNNRegressor(feature_columns=feature_cols,  
                                      hidden_units=[50, 50],
                                      model_dir='model_1')
```

```python
def wx_input_fn(X, y=None, num_epochs=None, shuffle=True, batch_size=50):  
    return tf.estimator.inputs.pandas_input_fn(x=X,
                                               y=y,
                                               num_epochs=num_epochs,
                                               shuffle=shuffle,
                                               batch_size=batch_size)
```

```python
wx_input_fn(X, y=None, num_epochs=None, shuffle=True, batch_size=50)
```

```python
evaluations = [] 
```

```python
STEPS = 100
```

```python
for i in range(100):  
    regressor.train(input_fn=wx_input_fn(X_train, y=y_train), steps=STEPS)
    evaluations.append(regressor.evaluate(input_fn=wx_input_fn(X_val, y_val, num_epochs=1, shuffle=False)))
```

```python
evaluations[0]
```

```python
pred = regressor.predict(input_fn=wx_input_fn(X_test, num_epochs=1,  shuffle=False))
```

```python
predictions = np.array([p['predictions'][0] for p in pred])
```

```python
print("The Explained Variance: %.2f" % explained_variance_score(  
                                            y_test, predictions))  
print("The Mean Absolute Error: %.2f " % mean_absolute_error(  
                                            y_test, predictions))  
print("The Median Absolute Error: %.2f " % median_absolute_error(  
                                            y_test, predictions))
```

# MAPE

```python
def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
mean_absolute_percentage_error(y_test,predictions)
```

# RESE

```python
def rmse(predictions, targets):
    return np.sqrt(((predictions - targets) ** 2).mean())
```

```python
rmse(y_test,predictions)
```
